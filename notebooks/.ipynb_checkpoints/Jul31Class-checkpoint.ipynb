```python
# Removal of an item from a list with if and while

primes = [2, 3, 5, 7, 11, 11, 13, 5, 15, 17]

#print(primes)

if 15 in primes:
#    print("15 is not supposed to be a prime!")
    primes.remove(15)

primes.remove(11)
primes.remove(5)
#print(primes)

colors = ["white", "blue", "yellow", "red", "brown", "blue", "blue", "bat", "bat", "bat"]
colors.remove("blue")
colors.remove("blue")

#if "bat" in colors:
#    colors.remove("bat")
    
while "bat" in colors:
    print("Removing bat")
    colors.remove("bat")
    
print(colors)
```

```python
# More list functions

# Clear list with .clear()

items = ["a", "b", "c", "b"]

#items.clear()

#print(items)

# Count occurence of an item x with .count(x)
# print(items.count("b"))

while items.count("b") > 1:
    items.remove("b")

print(items)

# Remove the item from a specified index with .pop([i])
first_item = items.pop(0)

print(first_item)
```

```python
# Deterministic vs non-deterministic loop

# While loop is non-deterministic loop because we do not know until at run-time when
# the loop will terminate.
# For example, the number guessing game, where it is non-deterministic loop because 
# we do not know ahead of time when the loop will end.
# while condition:
#    code_block
#    something to break the condition


# for loop on the other hand is a deterministic loop because we know 
# when the loop will end at the time of writing
# here we know that the loop will run for whatever is the length of the list
# for item in some_list:
    # some code_block
    
# While loops are superior in capability but for loops are simpler!

students = ["Rekha", "Anu", "Jay", "John", "Kaushik"]

#for student in students:
#    print(student)

i = 0
while i < len(students):
    print(students[i])
    i += 1
```

```python
# for loop hiegene
numbers = [9, 10, 11]

cubes = []

for n in numbers:
    cubes.append(n**3)

del n # best practice!
print(cubes)
# print(n) # the loop variable survives outside of loop
```
