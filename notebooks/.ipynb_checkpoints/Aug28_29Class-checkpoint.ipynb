```python
# Python Dictionaries!!

# A dictionary is a collection of key:value pairs. key => unique value 
# => there cannot be more than one key in a dictionary!!
```

```python
days = {"mon":"Monday", # dictionary items separated by comma
        "tue":"Tuesday",
        "wed":"Wednesday",
        "thu":"Thursday",
        "fri":"Friday"
       } # curly brackets like in sets except that the contents have a ':'

print(days)
```

```python
days = {"mon":"Monday", # dictionary items separated by comma
        "tue":"Tuesday",
        "wed":"Wednesday",
        "thu":"Thursday",
        "fri":"Friday"
       } # curly brackets like in sets except that the contents have a ':'

print(days["tue"]) # note the square bracket!

#print(days["sat"]) # missing key error

#print(days["Tuesday"]) # one-way, only access value by key and not the other way

# Add items to a dictionary
days["sat"] = "Saturday"
days["sun"] = "Sunday"

#print(days)
print(type(days))
print(len(days))
```

```python
# Find the item in a dictionary with highest value

scores = {
    "Megha":89,
    "Amit":87,
    "Varsha":79,
    "Manav":87,
    "Rohit":98,
    "Vishal":71
}

high_score = 0
topper = ""
for (name, score) in scores.items():
    if score > high_score:
        high_score = score
        topper = name

print("The topper is {} with a score of {}.".format(topper, high_score))
```

```python
# Problem: Count occurence of words in a sentence

# we need a dictionary with words as keys and their count as values.

sentence = "He had been watching me it seems from folds of ground and from \
amongst reed beds for nearly half a day wondering what I might be at I \
had come ashore to wander on the plain I like to be alone sometimes My \
ship was anchored in a bight of this deserted coast a good many miles \
away too many to walk back in the dark for a stranger like me"

# the:2, cat:1, sat:1, on:1, mat:1

words = sentence.split()
count = {} 

for w in words: # iterate over the words
    if w in count: # check if the word is already in dictionary
        count[w] += 1
    else: # if the word is not in dictionary then introduce it to dictionary
        count[w] = 1
        
#print(count)

max_w = ""
max_c = 0
for (w, c) in count.items():
    if c > max_c:
        max_c = c
        max_w = w

print("The word {} appears maximum times at {}".format(max_w, max_c))
```

```python
# A few more functions available with dictionaries

shapes = {
    "circle":   "A perfectly round shape",
    "square":   "A shape with four equal sides",
    "triangle": "A shape with three sides and three angles"
}

# the pop and popitem function
print(shapes.pop("square"))
print(shapes)

print(shapes.popitem())
print(shapes.popitem())
print(shapes.popitem()) # this will not work because the dictionary is empty!!
```

```python
shapes = {
    "circle":   "A perfectly round shape",
    "square":   "A shape with four equal sides",
    "triangle": "A shape with three sides and three angles"
}

more_shapes = {
    "rectangle":"a polygon with four sides",
    "parallelogram":"A polygon with four sides and opposite sides parallel"
}

shapes.update(more_shapes)

print(shapes)
```

```python
# creating dictionaries from files

# Open the file

cities = open("Cities.txt", "r")
d_cities = {}

for line in cities:
    l = line.split()
    d_cities[l[0]] = l[1]
    
cities.close()

print(d_cities)
```

```python
# Write to a file from dictionary
shapes = {
    "circle":   "A perfectly round shape",
    "square":   "A shape with four equal sides",
    "triangle": "A shape with three sides and three angles"
}

# open a file for writing purposes

so = open("shapes.txt", "w")

for k, v in shapes.items():
    so.write(k + " ")
    so.write(v + "\n")


so.close()
```

```python
# Nested dictionaries

# Dictionaries  inside of a dictionary!!

cities = {
    "Paris": {"country":"France", "pop":"4m"},
    "London": {"country":"England", "pop":"7m"},
    "Mumbai":{"country":"India", "pop":"30m"}
}
#print(cities)

# iterate over a nested dictionary

for k, v in cities.items():
    print(k)
    for k2, v2 in v.items():
        print("    " + k2, v2)
```
